```python
import matplotlib.pyplot as plt
```

```python
df.plot()
```

```python
quantity_per_year = df.groupby('Year')['Quantity'].sum()
quantity_per_year = quantity_per_year[quantity_per_year.index != 2017]

```

```python
plt.figure(figsize=(10, 5))
plt.plot(quantity_per_year.index, quantity_per_year.values, marker='o',linestyle='-')
plt.title('Total Car Sales in Norway by Year')
plt.xlabel('Year')      
plt.ylabel('Total Quantity')    
plt.grid()
plt.tight_layout()
```

```python
import_per_year = df.groupby("Year")["Import"].sum()
import_per_year = import_per_year[import_per_year.index != 2017]
plt.figure(figsize=(12,6))
plt.plot(import_per_year.index,import_per_year.values,linestyle='--',marker = 'o')
plt.xlabel('Year')
plt.ylabel('Total Import')  
plt.title("Total Import of Cars in Norway by Year")
plt.grid(True)
plt.tight_layout()
plt.savefig('../plots/import_per_year.png', dpi=300)
```

```python
avarage_C02 = df.groupby('Year')['Avg_CO2'].mean()
```

```python
df.columns
```

```python
avarage_C02= avarage_C02[avarage_C02.index != 2017]
```

```python
plt.figure(figsize=(12,6))
plt.plot(avarage_C02.index,
         avarage_C02.values,
         marker='o',
         linestyle='-',
         color ='orange',
         markerfacecolor='blue',
         markersize=10,
         alpha=1)
plt.xlabel("Year")
plt.ylabel("Avarage CO2")
plt.title("Avarage CO2 Emmission in Norway By Year")
plt.tight_layout()
plt.grid(True)
plt.savefig('../plots/avarage_C02.png')
```

```python
df['Date'] =pd.to_datetime(df[['Year','Month']].assign(Day=1))
import_per_month = df.groupby('Date')['Import'].sum()
```

```python
plt.figure(figsize=(12,6))
plt.plot(import_per_month.index,
         import_per_month.values,
         linestyle='-',marker = 'o',
         markeredgecolor = 'red',
         markerfacecolor='black',
         markersize=8,
         alpha=1,
         linewidth= 3)
plt.xlabel('Year')
plt.ylabel("Total import")
plt.title('Total imoport of cars in Norway by Month')
plt.tight_layout()
plt.grid(True)
plt.savefig('../plots/import_per_month.png', dpi=300)
```
