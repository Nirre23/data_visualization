```python
import pandas as pd

df = pd.read_csv('../data/steam_games.csv')
```

```python
df.head()
```

```python
# RÃ¤kna antal genre-taggar per spel
df['Tag Count'] = df['Genre Tags'].str.split(',').apply(lambda x: len(x) if isinstance(x, list) else 0)
```

```python
import matplotlib.pyplot as plt
fig , ax = plt.subplots(figsize=(10, 6))
ax.hist(df['Tag Count'])
plt.grid()
ax.set_title('Histogram of number of games with certain number of tags\non Steam among top 100')
ax.set_xlabel('Number tags per game')
ax.set_ylabel('Number of games')
plt.savefig('../plots/histogram.png', dpi=300, bbox_inches='tight')
```
