```python
import pandas as pd     
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../data/netflix_titles.csv')
df
```

```python
df.columns
```

```python
types = df['type'].value_counts()
```

```python
fig ,ax = plt.subplots(figsize=(3, 3))
ax.bar(types.index, types.values, color = 'steelblue',width=.2)
ax.set_title('Netflix Titles by Type', fontsize=14)
ax.set_xlabel('Type', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.legend(['Movies', 'TV Shows'], loc='upper right')
ax.
```
