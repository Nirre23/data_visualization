```python
import pandas as pd

df = pd.read_csv("../data/pokemon.csv")
```

```python
top_10_strongest = (df[["name","attack"]]).sort_values(by = "attack",ascending = False).head(10)
top_10_strongest
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.plot(top_10_strongest['name'],top_10_strongest['attack'],color = 'red',linewidth=2)
plt.scatter(top_10_strongest['name'],top_10_strongest['attack'])

plt.title("Top 10 strongest Pokemons",loc = 'left',pad= 15,fontsize = 15,color = 'gray')
plt.xlabel("Pokemon name",labelpad = 15,fontsize =15,color = 'gray')
plt.ylabel("Attack damage",labelpad= 15,fontsize = 15,color = 'gray')

ax = plt.gca()
ax.spines[['right','top']].set_visible(False)

plt.tick_params(axis="both",length = 0)


plt.tight_layout()
plt.savefig("../plots/strongest_pokemons.png",dpi = 300)
```
