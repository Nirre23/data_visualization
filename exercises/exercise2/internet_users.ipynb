```python
import pycountry_convert as pc

# Funktion som mappar ett land till en kontinent
def country_to_continent(country_name):
    try:
        # Konvertera från land till ISO alpha2 kod
        country_code = pc.country_name_to_country_alpha2(country_name)
        # Konvertera från alpha2 kod till kontinentkod
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        # Översätt kontinentkod till namn
        continent_name = {
            "AF": "Africa",
            "NA": "North America",
            "OC": "Oceania",
            "AN": "Antarctica",
            "AS": "Asia",
            "EU": "Europe",
            "SA": "South America"
        }
        return continent_name[continent_code]
    except Exception as e:
        return "Unknown"

# Lägg till en kolumn i din df
df["Continent"] = df["Entity"].apply(country_to_continent)

# Visa resultat
print(df[["Entity", "Continent"]].head())

# Se om några är Unknown
missing = df[df["Continent"] == "Unknown"]["Entity"].unique()
print("Saknade länder:", missing)
```

```python
# Specialmappning för saknade
manual_mapping = {
    "Cote d'Ivoire": "Africa",
    "Curacao": "North America",
    "Democratic Republic of Congo": "Africa",
    "East Timor": "Asia",
    "Faeroe Islands": "Europe",
    "Kosovo": "Europe",
    "Micronesia (country)": "Oceania",
    
    # Regioner och grupper
    "East Asia and Pacific (WB)": "Asia",
    "Europe and Central Asia (WB)": "Europe",
    "European Union (27)": "Europe",
    "High-income countries": "Europe",  # Eller skapa en egen kategori
    "Low-income countries": "Africa",
    "Lower-middle-income countries": "Asia",
    "Middle East and North Africa (WB)": "Africa",
    "Middle-income countries": "Asia",
    "North America (WB)": "North America",
    "South Asia (WB)": "Asia",
    "Sub-Saharan Africa (WB)": "Africa",
    "Upper-middle-income countries": "Europe",
    "Latin America and Caribbean (WB)": "South America",
    "World": "All Continents"  # Eller något annat du vill
}

# Fyll i där Continent är "Unknown"
df["Continent"] = df.apply(
    lambda row: manual_mapping[row["Entity"]] if row["Continent"] == "Unknown" and row["Entity"] in manual_mapping else row["Continent"],
    axis=1
)

# Kontrollera igen
print(df[df["Continent"] == "Unknown"])
```

```python
# Byt kolumnnamn för enkelhet
df = df.rename(columns={"Individuals using the Internet (% of population)": "Internet_Pct"})

# Grupp efter kontinent och år
continent_group = df.groupby(["Continent", "Year"])["Internet_Pct"].mean().reset_index()

print(continent_group.head())
```

```python
df_2023 = df[df["Year"] == 2023]
df_continent_2023 = df_2023.groupby("Continent")["Internet_Pct"].mean().reset_index()
df_continent_2023 = df_continent_2023.sort_values("Internet_Pct", ascending=True)
```

```python
fig = px.bar(
    df_continent_2023,
    y = 'Continent',
    x = 'Internet_Pct',
    text="Internet_Pct",  # etiketter direkt på staplarna
    color="Continent",  # olika färger per kontinent
    color_discrete_sequence=px.colors.qualitative.Set2,  # lite snyggare färger
)
fig.show()
```
