```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Sortera efter 'Ladder score' i fallande ordning
df_sorted = df.sort_values(by='Ladder_Score', ascending=False)

# VÃ¤lj de 10 fÃ¶rsta lÃ¤nderna
top_10_happy_countries = df_sorted.head(10)

# Skapa barplot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Country', y='Ladder_Score', data=top_10_happy_countries, palette='pastel')

# Ã„ndra fÃ¤rg pÃ¥ baren vid index 3 (fjÃ¤rde baren)
bars = ax.patches
bars[3].set_facecolor('Blue')  # Byt ut 'red' mot den fÃ¤rg du vill ha

ax.set_title("Sweden is the 4th happiest country in the world",
             pad=20, fontsize=16, color='black',loc = 'left')
ax.set_xlabel("Countries", fontsize=14, color='black',loc = 'left')
ax.set_ylabel("Happines Score", fontsize=14, color='black')
ax.xaxis.labelpad = 20
ax.yaxis.labelpad = 20

ax.spines[['top','right']].set_visible(False)  # DÃ¶ljer de Ã¶vre och hÃ¶gra spines
# Rotera x-axelns etiketter fÃ¶r bÃ¤ttre lÃ¤sbarhet
plt.xticks(rotation=0, ha='right')
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom= False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom= True
)
plt.xticks(ha = 'right')
bars = ax.patches
for i, bar in enumerate(bars):
    if i != 3:  # om det inte Ã¤r baren du vill highlighta
        bar.set_facecolor('gray'),bar.set_edgecolor('black'),bar.set_linewidth(1.5)
for label in ax.get_xticklabels():
    label.set_horizontalalignment('center')  # eller 'right' / 'center'
    label.set_x(label.get_position()[0] - 10000)
ax.annotate(text ='', xy=(3, 7.5), xytext=(3, 8.2),
            arrowprops=dict(arrowstyle='simple', color='black', lw=1), fontsize=12, color='black')
# Visa plot
plt.tight_layout()

plt.savefig('../plots/happiest_countries.png', dpi=100, bbox_inches='tight')
plt.show()
```
