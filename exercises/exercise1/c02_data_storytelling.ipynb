```python
import duckdb
import pandas as pd

df = pd.read_csv('../data/co2_annmean_mlo.csv', skiprows=43)  # Anpassa siffran!
df
```

```python
avg_co2 = df['mean'].groupby(df['year']).mean()
avg_co2
```

```python
import matplotlib.pyplot as plt
fig , ax = plt.subplots(figsize=(9, 6))
ax = avg_co2.plot(
    color = 'red',
    linewidth = 4,
)
ax.set_title("The annual mean of CO2 emmissions measured in Mauna Loa has\nincreased every year since 1959",
             loc= "left",family = "verdana",
             fontsize = 16,color = "black",pad= 20,fontweight = '50')

ax.set_xlabel("YEARS FROM 1959",loc= "left",family = "verdana", fontsize = 12,color = "#333333",labelpad= 10)

ax.set_ylabel("CO2 MOLE FRACTION IN PPM",loc= "top",family = "verdana", fontsize = 12,color = "#333333",labelpad= 10)

ax.spines[['top', 'right']].set_visible(False)

plt.savefig('../plots/co2_emissions.png', dpi=300, bbox_inches='tight')
```
