```python
import pandas as pd
import duckdb
import matplotlib.pyplot as plt

df = pd.read_csv("../data/norway_new_car_sales_by_month.csv")
df = duckdb.query(
    """--sql
    SELECT avg_CO2, import,quantity, year, month
    FROM df

"""
).df()

df["date"] = pd.to_datetime(
    df["Year"].astype(str) + "-" + df["Month"].astype(str).str.zfill(2), format="%Y-%m"
)

df = df.set_index("date")

ax = df["Avg_CO2"].plot()
```

```python
fig , ax = plt.subplots(figsize=(12, 6))
ax = df["Avg_CO2"].plot()

ax.spines[["top","right"]].set_visible(False)
ax.set_title("Average CO2 Emissions in Norway", fontsize=16,loc = "left",pad = 30,fontweight = "bold")
ax.set_ylabel("Average CO2 Emissions (g/km)", fontsize=14)
ax.set_xlabel("Year", fontsize=14,fontweight = "bold")
```

```python
def avg_co2_plot():   
        avg_co2 = df["Avg_CO2"]
        fig , ax = plt.subplots(figsize=(12, 6))


        ax = avg_co2.plot(color = "#133337")
        ax.spines[["top","right"]].set_visible(False)
        ax.spines[["left", "bottom"]].set_color("Gray"),ax.spines[["left", "bottom"]]
        ax.tick_params(axis='both', colors="gray",labelsize=9,length= 0,pad = 5)
        ax.tick_params(axis = 'y' ,pad = 7)

        ax.set_title("Groundbreaking Climate Success",
                    fontsize=16,
                    loc = "left",
                    pad = 30,
                    fontweight = "bold",
                    family = "monospace",
                    color = "gray"
        )
        ax.set_ylabel("Average CO2 Emissions (g/km)",
                    fontsize=14,
                    fontweight = "bold",
                    family = "monospace",labelpad= 20
                    ,color = "gray"
                    
        )

        ax.set_xlabel("Year",
                    fontsize=14,
                    fontweight = "bold",
                    family = "monospace",labelpad = 20 ,
                    color = "gray"
        )

        ax.text(
            0.2, 0.3,  # textens position (justera som du vill)
            "44% decrease in CO₂ emissions\nfrom 2007 to 2017",
            fontsize=12,
            color="#133337",
            fontweight="bold",
            ha="center",
            va="center",
            bbox=dict(boxstyle="round,pad=0.3", edgecolor="#133337", facecolor="lightgreen"),
            transform=ax.transAxes  # axekoordinater
        )
        plt.savefig("../plots/avg_co2_plot.png", dpi=300, bbox_inches='tight')
        plt.show()
        
avg_co2_plot()


```
