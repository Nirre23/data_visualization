```python
import pandas as pd
import duckdb
import matplotlib.pyplot as plt

df = pd.read_csv("../data/norway_new_car_sales_by_month.csv")
df = duckdb.query(
    """--sql
    SELECT avg_CO2, import,quantity, year, month
    FROM df

"""
).df()

df["date"] = pd.to_datetime(
    df["Year"].astype(str) + "-" + df["Month"].astype(str).str.zfill(2), format="%Y-%m"
)

df = df.set_index("date")

ax = df["Avg_CO2"].plot()
```

```python
fig , ax = plt.subplots(figsize=(12, 6))
ax = df["Avg_CO2"].plot()

ax.spines[["top","right"]].set_visible(False)
ax.set_title("Average CO2 Emissions in Norway", fontsize=16,loc = "left",pad = 30,fontweight = "bold")
ax.set_ylabel("Average CO2 Emissions (g/km)", fontsize=14)
ax.set_xlabel("Year", fontsize=14,fontweight = "bold")
```

```python
avg_co2 = df["Avg_CO2"]
fig , ax = plt.subplots(figsize=(12, 6))


ax = avg_co2.plot(color = "#133337",marker = "o",markersize = 5,linestyle = "-")
ax.spines[["top","right"]].set_visible(False)
ax.set_title("Avarage CO2 Emissions in Norway",
             fontsize=16,
             loc = "left",
             pad = 30,
             fontweight = "bold",
             family = "monospace"
)
ax.set_ylabel("Average CO2 Emissions (g/km)",
              fontsize=14,
              fontweight = "bold",
              family = "monospace"
)

ax.set_xlabel("Year",
               fontsize=14,
               fontweight = "bold",
               family = "monospace",labelpad = 20 
)

ax.text(
    0.2, 0.3,  # textens position (justera som du vill)
    "44% decrease in CO emissions\nfrom 2007 to 2017",
    fontsize=12,
    color="#133337",
    fontweight="bold",
    ha="center",
    va="center",
    bbox=dict(boxstyle="round,pad=0.3", edgecolor="#133337", facecolor="lightgreen"),
    transform=ax.transAxes  # axekoordinater
)


ax.annotate(
    "",                    # ingen text
    xy=(0.4, 0.2),         # pilspets
    xytext=(0.4, 0.35),     # dr pilen startar (samma som textens y)
    xycoords="axes fraction",
    arrowprops=dict(
        arrowstyle="fancy",
        color="black",
        linewidth=4,edgecolor="black",
    )
)
# Pilar rakt ner
ax.annotate(
    "",                    # ingen text
    xy=(0.4, 0.2),         # pilspets
    xytext=(0.4, 0.35),     # dr pilen startar (samma som textens y)
    xycoords="axes fraction",
    arrowprops=dict(
        arrowstyle="fancy",
        color="lightgreen",
        linewidth=2.5,edgecolor="black",
    )
)
```
